```python
import yasa
import numpy as np
import pandas as pd
import os, sys

# Output spindle files
output_path = "/home/birdfly/Projects/yasa/notebooks/result/"
```

```python
# Load data
f = np.load('2010201.npz')
data, chan = f['data'], f['chan']
sf = 250.
times = np.arange(data.size) / sf

# Transfer from array to dataframe
data = pd.DataFrame(np.transpose(data), columns = chan)

# Delete EOG1 EOG2 EMG1 EMG2 EMG3
data = data.drop(columns=['EOG1', 'EOG2', 'EMG1', 'EMG2', 'EMG3'])
chan = data.columns.to_numpy()

# transfer from dataframe to array and transpose
data = np.transpose(data.to_numpy())
```

```python
def save_spindle_data(index, row, data, sf):
    global output_path
    
    transpose_data = np.transpose(data)
    start = int(row['Start'] * sf)
    end = int(row['End'] * sf + 1)
    spindle_data = pd.DataFrame(transpose_data[start:end])
    if not os.path.exists(output_path):
        os.makedirs(output_path, mode=0o777) 
    pd.DataFrame(spindle_data).to_csv (output_path + "/" + row['Channel']+ str(index) + ".csv", index = None, mode = 'w',header=None)
```

```python
# Detect spindles signal
sp = yasa.spindles_detect_multi(data, sf, ch_names=chan, multi_only=False)
print(sp.shape[0], 'spindles detected.')
sp.head().round(3)
```

```python
for index, row in sp.iterrows():
    save_spindle_data(index, row, data, sf)
```
